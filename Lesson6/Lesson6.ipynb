### 1.	Задание (теорема сложения)
Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

**Ответ**<br>
2/6 (вероятность выпадения 2 (1/6) + вероятность выпадения 5 (1/6))

### 2.	Задание (теорема умножения)
Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

**Ответ**<br>
1/36 (вероятность выпадения 2 (1/6) * вероятность выпадения 5 (1/6))

### 3.	Задание 
Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости.

**Ответ**<br>
1/6 (подходят: выпадение 2, потом 5 или выпадение 5, потом 2)

### 4.	Задание  (Геометрическая вероятность + интервалы)
На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

**Решение**<br>
Точка С должна попать на отрезок [A, A+9] и отрезок [B-15, B]. т.е. на отрезок длиной 4см [A+5, A+9]. Вероятность попасть на отрезок длиной 4см внутри отрезка длиной 20см = 4/20 = 1/5<br>
**Ответ**<br>
Вероятность = 1/5

*Тут если быстро решать, то понятно, а если задумываться, то интуитивно есть возражение: почему вероятность попасть на отрезок [5,9] такая же, как, например, попасть на отрезок [13,17]? И в целом на большом отрезке в 20см есть множество подотрезков длиной 4см, а отрезок [5,9] такой один.*

### 6.	Задание. 
Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?

**Ответ**<br>
Чтобы наверняка найти номер, нужно сделать полный перебор - это число размещений из 9 по 2. А из 9 по 2 = 9!/7! = 72<br>
Вероятность угадать номер с первого раза: 1/72

### 7.	Задание** (необязательное) 
Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью все грани этого куба будут белыми?

**Решение**<br>
Среди полученных 27 кубиков:
* 8 - с тремя белыми гранями (угловые)
* 12 - с двумя белыми гранями
* 6 - с одной белой гранью
* 1 - черный

Чтобы собрать белый куб заново:
* Угловой кубик можно поставить любой из 8 на любое уголовое место, т.е. 8! способами (число перестановок 8ми элементов). При этом кубик на одном и том же месте можно повернуть 3мя способами, т.е. 3^8 для 8ми кубиков. Общее число возможных вариантов составить такие кубики = 3^8 * 8!
* Кубики с двумя белыми гранями можно поставить на любое из 12 мест, т.е. 12! способами. При этом кубик на одном и том же месте еще можно зеркально повернуть (2мя способами). Общее число возможных вариантов составить такие кубики = 2^12 * 12!
* Кубик с одной белой гранью можно поставить на любую из 6 позиций, и на одном месте есть 4 варианта вращения кубика. Общее число возможных вариантов составить такие кубики = 4^6 * 6!
* Черный кубик ставится только на 1 место, но его можно повернуть как угодно, т.е. 6 вариантов.

Всего возможных способов собрать куб:<br>
27! (число перестановок 27 кубиков) * 6^27 (каждый кубик можно поставить 6 способами)<br><br>
**Т.о. искомая вероятность:**<br>
(3^8 * 8! * 2^12 * 12! * 4^6 * 6! * 6) / (27! * 6^27)
