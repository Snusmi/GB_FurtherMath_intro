**2. Расчет длины вектора**

```python
import math

def len_v(v:list):
    if len(v)>1:
        try:
            new_list = [number**2 for number in v]
            sum = math.fsum(new_list)
            return math.fabs(math.sqrt(sum))
        except TypeError:
            print("Oops. That was not a valid vector")
    else:
        print("Unable to compute the length")
```

```python
a = [10,10,10]
b = [0,0,-10]
```

```python
print(len_v(a))
print(len_v(b))
```

### 2 Задание<br/>
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt

x = np.linspace(-5,5,21)
y1 = 3*x + 1
y2 = -(1/3)*x + 1

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

**Ответ: Из-за соотношения сторон/масштаба отображаемой области. Если отображаемую область ограничить одинаковыми отрезками по x и y, прямые будут выглядеть перпендикулярно.**

```python
x = np.linspace(-5,5,21)
y1 = 3*x + 1
y2 = -(1/3)*x + 1

fig = plt.figure()
ax = fig.add_subplot(111)
plt.xlim(-5,5)
plt.ylim(-5,5)
ax.set_xticks(np.arange(-5, 5, 2))
ax.set_yticks(np.arange(-5, 5, 2))
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```

* гиперболы<br/>

```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,1000)
r = 5
y1 = np.sqrt(r**2 - x**2)
y2 = -np.sqrt(r**2 - x**2)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y1)
ax.plot(x, y2)
ax.set_aspect('equal')
plt.show()
```

```python
x = np.linspace(-5,5,1000)
a = 5
b = 3
y1 = np.sqrt(b**2 - (b*x/a)**2)
y2 = -np.sqrt(b**2 - (b*x/a)**2)


fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y1)
ax.plot(x, y2)
ax.set_aspect('equal')
plt.show()
```

### 5. Задание (в программе))<br/>
1. Нарисуйте трехмерный график двух параллельных плоскостей.)<br/>
2. Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
import pylab
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = pylab.figure()
ax=Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 = 2*X + 2*Y
Z2 = 2*X + 2*Y + 20
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
plt.show()
```

```python
fig = pylab.figure()
ax=Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 + 3*Y**2
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
fig = pylab.figure()
ax=Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 = np.sqrt(X**2 + Y**2)
Z2 = -np.sqrt(X**2 + Y**2)
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
plt.show()
```
